## 1. Reading & Cleaning Data

```python
import pandas as pd
import numpy as np
```

```python
# reading data
unwto_df = pd.read_parquet('https://faculty.cs.niu.edu/~dakoop/cs640-2023sp/a4/unwto.parquet')
wb_df = pd.read_parquet('https://faculty.cs.niu.edu/~dakoop/cs640-2023sp/a4/wb.parquet')
oecd_df = pd.read_parquet('https://faculty.cs.niu.edu/~dakoop/cs640-2023sp/a4/oecd.parquet')
unpop_df = pd.read_parquet('https://faculty.cs.niu.edu/~dakoop/cs640-2023sp/a4/population.parquet')
```

```python
# cleaning oecd data
oecd_df['value'] = oecd_df['value'].replace('..', np.nan).astype(float)
oecd_df
```

```python
# converting values in WB dataset to million of US dollars
wb_df['value'] = wb_df['value']/ 1000000
wb_df
```

```python
# transforming wb data
pivoted_wb = wb_df.pivot(index = ['Country Name' ,'Country Code', 'Indicator Name', 'Year'], columns = 'Indicator Code', values = 'value')
pivoted_wb
```

```python
# transforming oecd data
pivoted_oecd = oecd_df.pivot(index = ['Country' ,'Year'], columns = 'Variable', values = 'value')
pivoted_oecd
```

```python
# transforming unwto data
pivoted_unwto = unwto_df.pivot( index = ['Country','CountryCode','Year'], columns = 'Variable', values = 'value')
pivoted_unwto
```

## 2. Data Imputation

```python
imputed_wb = pivoted_wb.groupby(['Country Name', 'Indicator Name'])[['ST.INT.RCPT.CD', 'ST.INT.XPND.CD']].apply(lambda x : x.interpolate(limit_area = 'inside'))
# grouped_wb = grouped_wb.interpolate()
imputed_wb
```

## 3. Schema Matching

```python
pivoted_unwto = pivoted_unwto.reset_index()
pivoted_unwto.rename(columns = {'Country': 'country', 'Year' : 'year', 'CountryCode': 'countryCodeNum', 'Inbound Tourism expenditure in the country' : 'receipts', 'Outbound Tourism expenditure in other countries' : 'expenditures' }, inplace = True)
pivoted_unwto.index.name = 'index'
```

```python
imputed_wb = imputed_wb.reset_index()
imputed_wb.rename(columns = {'Year' : 'year','Country Code' : 'countryCodeAlpha', 'Country Name' : 'country', 'ST.INT.RCPT.CD' : 'receipts', 'ST.INT.XPND.CD' : 'expenditures'}, inplace = True)
# imputed_wb
```

```python
pivoted_oecd = pivoted_oecd.reset_index()
pivoted_oecd.rename(columns = {'Country': 'country', 'Year' : 'year', 'Total international receipts' : 'receipts', 'Total international expenditure' : 'expenditures'}, inplace = True)
# pivoted_oecd
```

```python
# drop other columns
cols_list = ['countryCodeNum', 'countryCodeAlpha', 'country', 'year', 'receipts', 'expenditures']
datasets = [pivoted_unwto,imputed_wb, pivoted_oecd]
for dataset in datasets:
    for col in dataset.columns:
        if col not in cols_list:
            dataset.drop(columns = col, inplace = True)
```

```python
# adding source
pivoted_unwto = pivoted_unwto.assign(source = 'unwto')
imputed_wb = imputed_wb.assign(source = 'wb')
pivoted_oecd = pivoted_oecd.assign(source = 'oecd')
```

## 4. Entity Resolution

```python
country_codes = unpop_df[['Country', 'NumericCode', 'AlphaCode']].drop_duplicates()
# country_codes
```

## 4a. Using Country Codes

```python
# joining wb daatset with country_codes based on alpha codes
merged_wb = imputed_wb.merge(country_codes, how = 'left', left_on = 'countryCodeAlpha', right_on = 'AlphaCode')
merged_wb
```

```python
# drop other regions data in wb
merged_wb = merged_wb.dropna(subset = ['Country'])
# merged_wb
```

```python
# checking if all the codes are not empty
# merged_wb[merged_wb['AlphaCode'].isnull()]
```

```python
# dropping unnecessary columns and renaming to match the schema
final_wb = merged_wb.copy()
final_wb.drop(columns = ['Country', 'AlphaCode'], inplace = True)
```

```python
final_wb.rename(columns = {'NumericCode' : 'countryCodeNum'}, inplace = True)
final_wb = final_wb.drop_duplicates()
final_wb
```

```python
# joining unwto dataset with country_codes based on numeric codes
merged_unwto = pivoted_unwto.merge(country_codes, how = 'left', left_on = 'countryCodeNum', right_on = 'NumericCode')
merged_unwto = merged_unwto.drop_duplicates()
merged_unwto
```

```python
# resolving outdated codes using name matches
na_values = merged_unwto[merged_unwto['AlphaCode'].isnull()]
na_copy = na_values.copy()
na_copy.drop(columns = ['Country','NumericCode', 'AlphaCode'],inplace = True)
na_copy = na_copy.reset_index()
# na_copy
```

```python
na_copy['country'] = na_copy['country'].str.lower().str.replace('[^\w\s]', '', regex = True)
country_codes['Country'] = country_codes['Country'].str.lower().str.replace('[^\w\s]', '', regex = True)  
na_copy = na_copy.merge(country_codes, how = 'left', left_on = 'country', right_on = 'Country')
na_copy = na_copy.set_index('index')
# na_copy
```

```python
# updating modified data with the original data
merged_unwto.update(na_copy)
merged_unwto['year'] = merged_unwto['year'].astype(int)
# merged_unwto
```

```python
# checking which rows are null
# merged_unwto[merged_unwto['AlphaCode'].isnull()]
```

```python
# resolving remaining manually
code_map = {'saba' : 535, 'serbia and montenegro' : 688, 'sint eustatius' : 535}
alpha_code_map = {'saba' : 'BES', 'serbia and montenegro' : 'SRB', 'sint eustatius' : 'BES'}
merged_unwto.loc[merged_unwto['country'].isin(code_map.keys()), 'countryCodeNum'] = merged_unwto['country'].replace(code_map)
merged_unwto.loc[merged_unwto['country'].isin(alpha_code_map.keys()), 'AlphaCode'] = merged_unwto['country'].replace(alpha_code_map)
```

```python
# merged_unwto[merged_unwto['AlphaCode'].isnull()]
```

```python
final_unwto = merged_unwto.copy()
final_unwto.drop(columns = ['Country', 'NumericCode'], inplace = True)
final_unwto.rename(columns = {'AlphaCode' : 'countryCodeAlpha'}, inplace = True)
```

## 4b. Using name matches

```python
pivoted_oecd['country'] = pivoted_oecd['country'].str.lower().str.replace('[^\w\s]', '', regex = True)
country_codes['Country'] = country_codes['Country'].str.lower().str.replace('[^\w\s]', '', regex = True)  
```

```python
# joining oecd with country_codes based on name matches
merged_oecd = pivoted_oecd.merge(country_codes, how = 'left', left_on = 'country',right_on = 'Country')
merged_oecd
```

## 4d. Concatenation

```python
concat_df = pd.concat([final_wb, final_unwto, final_oecd])
```

```python
concat_df
```

## 5. Data Fusion

```python
final_df = concat_df.groupby(['year', 'countryCodeAlpha', 'countryCodeNum']).agg({'expenditures': 'mean', 'receipts': 'mean', 'country' : 'first'}).reset_index()
final_df = final_df.drop_duplicates()
final_df['countryCodeNum'] = final_df['countryCodeNum'].astype(int)
```

```python
# final_df[(final_df['countryCodeAlpha'] == 'USA') & (final_df['year'] >= 2012)]
```

## 6. Data Integration 

```python
integrated_df = pd.merge(final_df, unpop_df, how = 'left', left_on =['countryCodeAlpha', 'year'], right_on = ['AlphaCode', 'Year'])
```

```python
integrated_df = integrated_df.reset_index()
integrated_df.drop(columns = ['index','AlphaCode', 'Year', 'country', 'NumericCode'], inplace = True)
integrated_df
```

```python
integrated_df = integrated_df.drop_duplicates()
```

```python
# Calculate per capita receipts and expenditures
integrated_df = integrated_df.copy()
integrated_df.rename(columns = {'Population-Thousands': 'population'}, inplace = True)
integrated_df['perCapitaRcpts'] = integrated_df['receipts'] / (integrated_df['population'] /1000)
integrated_df['perCapitaExpnds'] = integrated_df['expenditures'] / (integrated_df['population'] / 1000)
```

```python
integrated_df = integrated_df.reindex(columns=['Country', 'countryCodeNum', 'countryCodeAlpha', 'year', 'receipts', 'expenditures', 'population', 'perCapitaRcpts', 'perCapitaExpnds'])
```
